### 결측치 제거
- 리뷰가 없는 경우 Review 값들이 결측치로 저장
- 리뷰가 없는 상품들을 제거하기 위해 전처리 진행

```python
import pandas as pd

df=pd.read_csv("lifestyle_reviews_finn.csv",encoding='utf-8-sig')
df.tail()
```

```python
# Review가 모두 결측치일 경우 해당 행 제거

# df3.iloc[:, 1:] = df3.iloc[:, 1:].copy()
# df3.iloc[:,1:].dropna(axis=0,how='all',inplace=True)
df.dropna(axis=0, how='all', subset=df.columns[1:], inplace=True) # subset 사용
df.tail()
```

```python
# 인덱스 재정렬
df.reset_index(drop=True,inplace=True)
df.tail()
```

```python
df.to_csv("../lifestyle_reviews_final.csv",encoding='utf-8-sig',index=False)
```

```python
# 결측행 제거 함수
def nan_processing(df_ex):
    df_ex.dropna(axis=0, how='all', subset=df_ex.columns[1:], inplace=True) # subset 사용
    df_ex.reset_index(drop=True,inplace=True)
```

```python
df2=pd.read_csv("jordan_reviews.csv",encoding='utf-8-sig')
df3=pd.read_csv("running_reviews.csv",encoding='utf-8-sig')
df_list=[df2,df3]
for i in df_list:
    nan_processing(i)

df2.to_csv("../jordan_reviews_final.csv",encoding='utf-8-sig',index=False)
df3.to_csv("../running_reviews_final.csv",encoding='utf-8-sig',index=False)
```

```python
basketball_reviews=pd.read_csv("basketball_reviews.csv",encoding='utf-8-sig')
gym_reviews=pd.read_csv("gym_reviews.csv",encoding='utf-8-sig')
soccer_reviews=pd.read_csv("soccer_reviews.csv",encoding='utf-8-sig')

df_list2=[
        ("basketball_reviews", basketball_reviews),
        ("gym_reviews", gym_reviews),
        ("soccer_reviews", soccer_reviews),
          ]
for name,df in df_list2:
    nan_processing(df)
    df.to_csv(f"../{name}_final.csv",encoding='utf-8-sig',index=False)
```
