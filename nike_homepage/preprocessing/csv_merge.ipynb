### 라이프스타일 카테고리 csv 병합
- 라이프스타일 카테고리 상품수가 많은 관계로 중간에 한번 끊어서 진행
- 따라서 두 파일간 병합이 필요해 병합을 진행

```python
import pandas as pd

df=pd.read_csv("lifestyle_reviews_fin.csv")
df.tail()
```

```python
# df의 마지막 상품 df2의 첫 상품과 중복이라 제거
df=df[:-1]
```

```python
df
```

```python
df2=pd.read_csv("lifestyle_reviews_fin2.csv")
df2.head()
```

```python
df3=pd.concat([df,df2],axis=0)
df3
```

```python
# 인덱스 재설정
df3.reset_index(drop=True,inplace=True)
```

```python
df3.tail()
```

```python
df3.to_csv("lifestyle_reviews_finn.csv",encoding="utf-8-sig",index=False)
```

```python
# Review가 모두 결측치일 경우 해당 행 제거

# df3.iloc[:, 1:] = df3.iloc[:, 1:].copy()
# df3.iloc[:,1:].dropna(axis=0,how='all',inplace=True)
df3.dropna(axis=0, how='all', subset=df3.columns[1:], inplace=True) # subset 사용
df3.tail()
```

```python
# 인덱스 재정렬
df3.reset_index(drop=True,inplace=True)
```

```python
df3.tail()
```

```python
df3.to_csv("../lifestyle_reviews.csv",encoding='utf-8-sig',index=False)
```
